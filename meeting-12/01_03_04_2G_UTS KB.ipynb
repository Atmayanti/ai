### Keterangan Kriteria
- ICS01 : biji hitam
- ICS02 : biji hitam sebahagian
- ICS03 : biji hitam pecah
- ICS04 : gelondong
- ICS05 : biji cokelat
- ICS06 : kulit kopi (husk) ukuran besar
- ICS07 : kulit kopi (husk) ukuran sedang
- ICS08 : kulit kopi (husk) ukuran kecil
- ICS09 : biji kulit tanduk
- ICS10 : kulit tanduk ukuran besar
- ICS11 : kulit tanduk ukuran sedang
- ICS12 : kulit tanduk ukuran kecil
- ICS13 : biji pecah
- ICS14 : biji muda
- ICS15 : biji berlubang satu
- ICS16 : biji berlubang lebih dari satu
- ICS17 : biji bertutul-tutul (untuk proses basah)
- ICS18 : ranting tanah atau batu berukuran besar
- ICS19 : ranting tanah atau batu berukuran sedang
- ICS20 : ranting tanah atau batu berukuran kecil

### Keterangan Rules sebagai berikut:
- Mutu 1 : nilai kecacatan maksimum 11
- Mutu 2 : nilai kecacatan 12 sampai 25
- Mutu 3 : nilai kecacatan 26 sampai 44
- Mutu 4a : nilai kecacatan 45 sampai 60
- Mutu 4b : nilai kecacatan 61 sampai 80
- Mutu 5 : nilai kecacatan 81 sampai 150
- Mutu 6 : nilai kecacatan 151 sampai 225

```python
# from operator import itemgetter

print('Silakan Isi Jumlah Kopi yang Memiliki Cacat Dibawah : ')

# himpunan kriteria kecacatan kopi berdasarkan SNI
list_criteria = [['ICS01','biji hitam',1], ['ICS02','biji hitam sebahagian',0.5], ['ICS03','biji hitam pecah',0.5],
                 ['ICS04','gelondong',1], ['ICS05','biji cokelat',0.25], ['ICS06','kulit kopi (husk) ukuran besar',1],
                 ['ICS07','kulit kopi (husk) ukuran sedang',0.5], ['ICS08','kulit kopi (husk) ukuran kecil',0.2], ['ICS09','biji kulit tanduk',0.5],
                 ['ICS10','kulit tanduk ukuran besar',0.5], ['ICS11','kulit tanduk ukuran sedang',0.2], ['ICS12','kulit tanduk ukuran kecil',0.1],
                 ['ICS13','biji pecah',0.2],['ICS14','biji muda',0.2],['ICS15','biji berlubang satu',0.1],['ICS16','biji berlubang lebih dari satu',0.2],
                 ['ICS17','biji bertutul-tutul (untuk proses basah)',0.1],['ICS18','ranting tanah atau batu berukuran besar',5],
                 ['ICS19','ranting tanah atau batu berukuran sedang',2],['ICS20','ranting tanah atau batu berukuran kecil',1]] 

# fakta
facts = []
trashs = []
seeds = []


# perulangan untuk mengecek apakah kecacatan termasuk berupa sampah
# atau kecacatan termasuk cacat pada biji
for i in range(len(list_criteria)):
    if list_criteria[i][0] == 'ICS18':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS19':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS20':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS06':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS07':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS11':
        trashs.append(list_criteria[i])
    elif list_criteria[i][0] == 'ICS08':
        trashs.append(list_criteria[i])
    else:
        seeds.append(list_criteria[i])
        

# sorting sampah dari nilai yang paling besar
sorted_trashs = sorted(trashs, key=itemgetter(2), reverse=True)


# sorting biji kopi dari nilai kecacatan yang paling besar
sorted_seeds = sorted(seeds, key=itemgetter(2), reverse=True)


# perulangan untuk mendapatkan fakta dari trash
for i in range(len(sorted_trashs)):
    ask = 'Masukkan jumlah kopi '+sorted_trashs[i][1]+': '
    val = float(input(ask))
    result = [sorted_trashs[i][0], sorted_trashs[i][2]*val]
    facts.append(result)
    
    
# perulangan untuk mendapatkan fakta dari seed
for i in range(len(sorted_seeds)):
    ask = 'Masukkan jumlah kopi '+sorted_seeds[i][1]+': '
    val = float(input(ask))
    result = [sorted_seeds[i][0], sorted_seeds[i][2]*val]
    facts.append(result)
    

# fungsi untuk menghitung nilai total kecacatan
def total(facts):
    total = 0
    for i in range(len(facts)):
        total += facts[i][1]
    return total


# fungsi untuk mengecek grade kopi berdasarkan nilai total kecacatan
def grade():
    sum = total(facts)
    if (sum < 11): 
        return '1'
    elif (sum <= 25) & (sum >= 12):
        return '2'
    elif (sum <= 44) & (sum >= 26):
        return '3'
    elif (sum <= 60) & (sum >= 45):
        return '4a'
    elif (sum <= 80) & (sum >= 61):
        return '4b'
    elif (sum <= 150) & (sum >= 81):
        return '5'
    elif (sum <= 225) & (sum >= 151):
        return '6'
    else:
        return 'n/a'
    

print()
print('Nilai total kecacatan : ',total(facts))
print('Grade : ',grade())

# Yang belum :
# - dataset
# - visualisasi matplotblib / seaborn
```
